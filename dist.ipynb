```python
plt.figure(figsize=(12,10))
plt.hist(df_dat.loc[:,'bedrooms'], bins = np.arange(-0.5, 7, 1), color = 'cornflowerblue', ec = "white")
plt.xticks(np.arange(0,7, 1))
plt.xticks(fontsize = 14)
plt.xlabel("Number of Bedrooms", fontsize = 18)
plt.yticks(fontsize = 14)
plt.ylabel("Frequency",fontsize = 18)
plt.title("Distribution of Number of Bedrooms in Example Data Set ", fontsize = 20)
plt.savefig('figures_for_printing/bedrooms-July18', bbox = 'tight')
plt.show()
```
